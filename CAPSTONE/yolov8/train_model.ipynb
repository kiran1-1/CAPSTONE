```python
import cv2
import numpy as np

# Load the input image
image_path = "/Users/kiran_g/Documents/Capstone/final project/Parking_YOLOV8/Results//Parking_cropped_result.jpg"  # Replace with your image file path
image = cv2.imread(image_path)

# Create a blank black mask with the same size as the input image
mask = np.zeros(image.shape[:2], dtype=np.uint8)

# Global variables for drawing rectangles
drawing = False
ix, iy = -1, -1

# Mouse callback function to draw rectangles
def draw_rectangle(event, x, y, flags, param):
    global ix, iy, drawing, mask

    if event == cv2.EVENT_LBUTTONDOWN:
        # Start drawing
        drawing = True
        ix, iy = x, y

    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            temp_mask = mask.copy()
            cv2.rectangle(temp_mask, (ix, iy), (x, y), 255, -1)
            temp_display = cv2.addWeighted(image, 0.7, cv2.merge([temp_mask]*3), 0.3, 0)
            cv2.imshow("Draw Mask", temp_display)

    elif event == cv2.EVENT_LBUTTONUP:
        # Finalize the rectangle
        drawing = False
        cv2.rectangle(mask, (ix, iy), (x, y), 255, -1)
        temp_display = cv2.addWeighted(image, 0.7, cv2.merge([mask]*3), 0.3, 0)
        cv2.imshow("Draw Mask", temp_display)

# Display the image for drawing
cv2.imshow("Draw Mask", image)
cv2.setMouseCallback("Draw Mask", draw_rectangle)

print("Draw parking areas by dragging the mouse. Press 'q' to save and quit.")

# Wait for user to finish drawing
while True:
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):  # Press 'q' to quit and save the mask
        break

# Save the mask
output_path = "/Users/kiran_g/Documents/Capstone/final project/Parking_YOLOV8/Results//MASK.jpg"
cv2.imwrite(output_path, mask)
print(f"Mask saved to {output_path}")

# Release resources
cv2.destroyAllWindows()
```

```python
import cv2

# Global variables for mouse callback
cropping = False
start_x, start_y, end_x, end_y = -1, -1, -1, -1

def mouse_crop(event, x, y, flags, param):
    """
    Mouse callback function to select cropping region.
    """
    global start_x, start_y, end_x, end_y, cropping

    if event == cv2.EVENT_LBUTTONDOWN:
        # Start drawing the rectangle
        cropping = True
        start_x, start_y = x, y

    elif event == cv2.EVENT_MOUSEMOVE:
        if cropping:
            end_x, end_y = x, y

    elif event == cv2.EVENT_LBUTTONUP:
        # Finish drawing the rectangle
        cropping = False
        end_x, end_y = x, y

def crop_video_and_mask(video_path, mask_path, output_video_path, output_mask_path):
    global start_x, start_y, end_x, end_y

    # Load the video and mask
    cap = cv2.VideoCapture(video_path)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

    if mask is None:
        print("Error: Could not load mask image. Check the file path.")
        return

    # Get the first frame of the video
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read video.")
        cap.release()
        return

    # Ensure the mask dimensions match the video frame size
    frame_height, frame_width = frame.shape[:2]
    if mask.shape != (frame_height, frame_width):
        mask = cv2.resize(mask, (frame_width, frame_height))

    # Display the video frame and set mouse callback
    cv2.imshow("Video Frame", frame)
    cv2.setMouseCallback("Video Frame", mouse_crop)

    print("Draw a cropping rectangle on the video frame using the mouse. Press 'q' to confirm.")
    while True:
        temp_frame = frame.copy()
        if start_x != -1 and start_y != -1 and end_x != -1 and end_y != -1:
            # Draw the rectangle on the frame
            cv2.rectangle(temp_frame, (start_x, start_y), (end_x, end_y), (0, 255, 0), 2)
        cv2.imshow("Video Frame", temp_frame)
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):  # Press 'q' to confirm cropping
            break
    cv2.destroyAllWindows()

    # Ensure valid coordinates
    x, y = min(start_x, end_x), min(start_y, end_y)
    w, h = abs(end_x - start_x), abs(end_y - start_y)

    # Crop the mask based on the selected region
    cropped_mask = mask[y:y+h, x:x+w]
    cv2.imwrite(output_mask_path, cropped_mask)
    print(f"Cropped mask saved to: {output_mask_path}")

    # Initialize the video writer for the cropped video
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    output_size = (w, h)
    cropped_video_writer = cv2.VideoWriter(output_video_path, fourcc, fps, output_size)

    # Process the video frame by frame
    cap.set(cv2.CAP_PROP_POS_FRAMES, 0)  # Reset video to the first frame
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Crop the frame
        cropped_frame = frame[y:y+h, x:x+w]

        # Write the cropped frame to the output video
        cropped_video_writer.write(cropped_frame)

    cap.release()
    cropped_video_writer.release()
    print(f"Cropped video saved to: {output_video_path}")

# Paths for the input video and mask
# Paths for the input video and mask
video_path = "/Users/kiran_g/Documents/Capstone/parking/parking_1920_1080.mp4"  # Replace with your video file path
mask_path = "/Users/kiran_g/Documents/Capstone/parking/mask_1920_1080.png"   # Replace with your mask file path

# Paths for the output cropped video and mask
output_video_path = "/Users/kiran_g/Documents/Capstone/parking/cropped_video.mp4"  # Path to save the cropped video
output_mask_path = "/Users/kiran_g/Documents/Capstone/parking/cropped_mask.png"  # Path to save the cropped mask

# Run the function
crop_video_and_mask(video_path, mask_path, output_video_path, output_mask_path)
```

```python
import cv2
import numpy as np

def draw_bounding_boxes_on_video(video_path, mask_path, output_video_path):
    """
    Detect bounding boxes from the mask and draw them on the video.
    
    Args:
    - video_path (str): Path to the input video file.
    - mask_path (str): Path to the mask image file.
    - output_video_path (str): Path to save the output video with bounding boxes.
    """
    # Load the mask
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    if mask is None:
        print("Error: Could not load mask image. Check the file path.")
        return

    # Find contours (white areas in the mask)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Get the bounding boxes for each white region
    bounding_boxes = [cv2.boundingRect(contour) for contour in contours]

    # Load the video
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video file.")
        return

    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')

    # Initialize the video writer
    output_video_writer = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

    print("Processing video and drawing bounding boxes...")
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Draw the bounding boxes on the current frame
        for box in bounding_boxes:
            x, y, w, h = box
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Write the frame with bounding boxes to the output video
        output_video_writer.write(frame)

    # Release resources
    cap.release()
    output_video_writer.release()
    print(f"Output video with bounding boxes saved to: {output_video_path}")

# Paths for the video and mask
video_path = "/Users/kiran_g/Documents/Capstone/parking/cropped_video.mp4"  # Path to save the cropped video
mask_path = "/Users/kiran_g/Documents/Capstone/parking/cropped_mask.png" 

output_video_path = "/Users/kiran_g/Documents/Capstone/parking/output_with_boxes.mp4"  # Path to save the output video

# Run the function
draw_bounding_boxes_on_video(video_path, mask_path, output_video_path)
```
